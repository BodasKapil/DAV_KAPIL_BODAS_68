```r


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print(titanic_data.head())

# Visualization 1: Countplot of Survived
sns.countplot(x='survived', data=titanic_data)
plt.title('Survival Count')
plt.show()

```

```r

# Visualization 2: Countplot of Survived based on Pclass
sns.countplot(x='survived', hue='pclass', data=titanic_data)
plt.title('Survival Count based on Pclass')
plt.show()

```

```r
# Visualization 3: Age Distribution
sns.histplot(titanic_data['age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution')
plt.show()
```

```r

# Visualization 4: Scatter plot of Fare vs Age
sns.scatterplot(x='age', y='fare', data=titanic_data)
plt.title('Scatter plot of Fare vs Age')
plt.show()
```

```r

# Visualization 5: Boxplot of Fare based on Pclass
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Boxplot of Fare based on Pclass')
plt.show()
```

```r
import pandas as pd
from plotnine import ggplot, aes, geom_bar, geom_point, geom_boxplot, labs, theme_minimal

# Load the tips dataset
tips_data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

# Visualization 1: Countplot of Day
plot1 = ggplot(tips_data, aes(x='day')) + geom_bar() + labs(title='Day-wise Tips Count') + theme_minimal()
print(plot1)

# Visualization 2: Scatter plot of Total Bill vs Tip
plot2 = ggplot(tips_data, aes(x='total_bill', y='tip')) + geom_point() + labs(title='Scatter plot of Total Bill vs Tip') + theme_minimal()
print(plot2)

# Visualization 3: Boxplot of Tip based on Day
plot3 = ggplot(tips_data, aes(x='day', y='tip')) + geom_boxplot() + labs(title='Boxplot of Tip based on Day') + theme_minimal()
print(plot3)
```

```r
# Install and load the necessary library (if not already installed)
if (!require(ggplot2)) {
  install.packages("ggplot2")
}
library(ggplot2)

# Load the tips dataset
tips_data <- read.csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Visualization 1: Countplot of Day
plot1 <- ggplot(tips_data, aes(x = day)) +
  geom_bar() +
  labs(title = "Day-wise Tips Count") +
  theme_minimal()
print(plot1)

# Visualization 2: Scatter plot of Total Bill vs Tip
plot2 <- ggplot(tips_data, aes(x = total_bill, y = tip)) +
  geom_point() +
  labs(title = "Scatter plot of Total Bill vs Tip") +
  theme_minimal()
print(plot2)

# Visualization 3: Boxplot of Tip based on Day
plot3 <- ggplot(tips_data, aes(x = day, y = tip)) +
  geom_boxplot() +
  labs(title = "Boxplot of Tip based on Day") +
  theme_minimal()
print(plot3)
```

```r
# Install necessary packages
install.packages("plotly")

# Load the required library
library(plotly)

# Create some sample data
data <- data.frame(
  x = c(1, 2, 3, 4, 5),
  y = c(10, 14, 18, 24, 30)
)

# Create a plot using plot_ly
plot <- plot_ly(data, x = ~x, y = ~y, type = 'scatter', mode = 'lines+markers', name = 'Line')

# Display the plot
print(plot)
```
