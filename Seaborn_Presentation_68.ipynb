# **Seaborn**

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
iris = pd.read_csv("/content/Iris.csv")
iris.head()
```

```python
iris["Species"].value_counts()
```

### 1. **barplot**: Shows relationship between categorical and numerical variables through the height of bars.

```python
sns.barplot(x='Species', y='SepalLengthCm', data=iris)
```

```python
sns.barplot(x='Species', y='SepalWidthCm', data=iris)
```

## 2. **Countplot**: Displays frequency of unique values in a categorical variable through bars.

```python
sns.countplot(x='SepalWidthCm', data=iris)
plt.xlabel('SepalWidthCm')
plt.ylabel('Count')
plt.title('Count Plot of Iris Species')
```

## **3. displot**: Plots univariate distribution of observations, providing a histogram with optional kernel density estimation or rug plot.

```python
sns.displot(iris, x="PetalWidthCm", kind='kde')
plt.title('Distribution of Sepal Length')
plt.xlabel('PetalWidthCm')
plt.ylabel('Count')
```

```python

sns.displot(data=iris, x='SepalWidthCm', kind='kde')

# Set plot title and labels
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Show the plot
plt.show()

```

### 4. **Scatterplot**: Displays relationship between two numerical variables through dots on a 2D plane.




```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=iris, hue='Species')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.grid(True)
```

### 5. **Jointplot**: Combines scatterplot with histograms to visualize relationship and distribution between two variables.

```python
#A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, hue='Species')
# sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, kind="scatter", hue="Species")
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.grid(True)
```

### 6.**Boxplot**: Illustrates distribution of numerical data through quartiles, outliers, and median in a concise manner.

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
plt.title('Boxplot')

plt.show()
```

## 7. **Stripplot**: Represents distribution of data points along a single axis, useful for visualizing categorical data with overlaps.

```python
# ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, edgecolor="gray",hue='Species')
```

## 8. **Density**: Visualizes the distribution of a single numerical variable through a smoothed density curve.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris, x="SepalLengthCm", hue="Species", fill=True, alpha=0.8)

plt.title('Density Plot of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.legend(title='Species')
plt.show()
```

## 9. **Violinplot**: Displays distribution of numerical data across various levels of a categorical variable using kernel density estimation and box plot

```python
sns.violinplot(x='Species', y='SepalLengthCm', data=iris)
plt.title(f'Violin Plot of SepalLengthCM by Species')
```

```python
sns.violinplot(x='Species', y='PetalLengthCm', data=iris)
plt.title(f'Violin Plot of PetalLengthCM by Species')
```

## 10. **Pairplot**: Generates grid of scatterplots for quick examination of pairwise relationships in a dataset.

```python
plt.figure(figsize=(10, 8))
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")#hist,auto,none)
plt.suptitle('Pair Plot of Iris Dataset Features')
```

## 11. **heatmap**: Visualizes matrix-like data through colored cells, useful for showcasing correlations or distributions.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Iris Dataset Features')
plt.show()
```

**12. stripplot**: Represents distribution of data points along a single axis, useful for visualizing categorical data with overlaps.


 In a strip plot, data points are distributed along a single axis based on categories, but they may overlap. To mitigate overlap, jittering (small random displacement) is often applied to spread out the points horizontally, making each data point visible.

```python
sns.stripplot(x='Species', y='PetalWidthCm', data=iris,jitter=True)
plt.xlabel('Species')
plt.ylabel('PetalWidthCm')
plt.title('Strip Plot of Petal Length for Iris Species')
```

## **13. swarmplot**: Represents each data point in a categorical scatterplot avoiding overlap, useful for small datasets. Data points are positioned so that they do not overlap with each other.

```python
sns.swarmplot(x='Species', y='PetalWidthCm', data=iris)
plt.xlabel('Species')
plt.ylabel('PetalWidthCm')
plt.title('Strip Plot of Petal Length for Iris Species')
```

## **14. regplot**: Displays a scatterplot with a regression line to show relationship between two numerical variables.

```python
sns.regplot(x="PetalWidthCm", y="PetalLengthCm", data=iris);
```

```python
import seaborn as sns

# Load Iris dataset
iris = sns.load_dataset('iris')

# Create a regression plot and retrieve the linear regression parameters
regression_result = sns.regplot(x="petal_width", y="petal_length", data=iris).get_lines()[0].get_data()

# Extract the intercept (b0) and slope (b1)
b0 = regression_result[1][0]
b1 = regression_result[1][-1] - b0


print("Intercept (b0):", b0)
print("Slope (b1):", b1)
print("Y=",b0,"+",b1,"x")
```
